```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('20231010-20231027_games.csv')
df = df.set_index('gameID')
```

```python
x = df['gameDate'].unique()
y = df['gameDate'].value_counts()[df['gameDate'].unique()]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Games per Date')
ax.set_xlabel('Date')
plt.xticks(rotation=45, ha='right')
plt.show()
```
